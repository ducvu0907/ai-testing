```python
def show(img):
    plt.imshow(img)
```

```python
show(cv2.resize(img, (50, 50)))
```

```python
kernel = np.ones((5, 5), np.uint8) 
show(cv2.erode(img, kernel))
```

```python
show(cv2.GaussianBlur(img, (7, 7), 0))
```

```python
# Image rotation parameter
center = (img.shape[1] // 2, img.shape[0] // 2)
angle = 30
scale = 1

# getRotationMatrix2D creates a matrix needed for transformation.
rotation_matrix = cv2.getRotationMatrix2D(center, angle, scale)
rotated_image = cv2.warpAffine(img, rotation_matrix, (img.shape[1], img.shape[0]))
```

```python
show(rotated_image)
```

```python
image_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

width = image_rgb.shape[1]
height = image_rgb.shape[0]

tx = 100
ty = 70

# Translation matrix
translation_matrix = np.array([[1, 0, tx], [0, 1, ty]], dtype=np.float32)
# warpAffine does appropriate shifting given the Translation matrix.
translated_image = cv2.warpAffine(image_rgb, translation_matrix, (width, height))
show(translated_image)
```

```python
edges = cv2.Canny(image= image_rgb, threshold1=100, threshold2=700)
show(edges)
```

```python
start_point = (0, 0)

# End coordinate, here (250, 250)
# represents the bottom right corner of image
end_point = (250, 250)

# Green color in BGR
color = (0, 255, 0)

# Line thickness of 9 px
thickness = 9

# Using cv2.line() method
# Draw a diagonal green line with thickness of 9 px
image = cv2.line(img, start_point, end_point, color, thickness)
show(image)
```
